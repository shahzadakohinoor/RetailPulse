```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
import pandas as pd

df = pd.read_csv("../data/raw/retail_data.csv")
df.head()
```

```python
df.shape
df.columns
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df["sales"], bins=30, kde=True)
plt.title("Sales Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="frequency", y="sales", data=df)
plt.title("Frequency vs Sales")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

### Insights:
- Higher frequency customers tend to generate more sales
- Recency impacts customer engagement
- No missing values in dataset
