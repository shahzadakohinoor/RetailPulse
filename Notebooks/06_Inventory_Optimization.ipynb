```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
    "product_id": range(1, 101),
    "demand": np.random.randint(50, 500, 100),
    "inventory": np.random.randint(20, 300, 100),
    "lead_time": np.random.randint(1, 10, 100)
})

df.head()
```

```python
df["safety_stock"] = df["demand"] * df["lead_time"] * 0.1
```

```python
df["reorder_point"] = (df["demand"] * df["lead_time"]) + df["safety_stock"]
```

```python
df["reorder_quantity"] = np.maximum(
    df["reorder_point"] - df["inventory"], 0
)
```

```python
plt.figure(figsize=(8,4))
plt.bar(df["product_id"][:20], df["reorder_quantity"][:20])
plt.title("Reorder Quantity (Top 20 Products)")
plt.xlabel("Product ID")
plt.ylabel("Reorder Quantity")
plt.show()
```

```python
df.to_csv("../data/processed/inventory_optimization.csv", index=False)
```

### Inventory Optimization Insights
- Reorder point helps avoid stockouts
- Safety stock ensures buffer during delays
- High demand products require frequent restocking
