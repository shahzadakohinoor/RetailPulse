```python
import pandas as pd

df = pd.read_csv("../data/raw/retail_data.csv")
df.head()
```

```python
df["R_score"] = pd.qcut(df["recency"], 4, labels=[4,3,2,1])
df["F_score"] = pd.qcut(df["frequency"], 4, labels=[1,2,3,4])
df["M_score"] = pd.qcut(df["sales"], 4, labels=[1,2,3,4])
```

```python
df["RFM_score"] = (
    df["R_score"].astype(str) +
    df["F_score"].astype(str) +
    df["M_score"].astype(str)
)
```

```python
def segment_customer(row):
    if row["RFM_score"] == "444":
        return "Best Customers"
    elif row["F_score"] == 4:
        return "Loyal Customers"
    elif row["R_score"] == 1:
        return "At Risk"
    else:
        return "Normal"

df["segment"] = df.apply(segment_customer, axis=1)
```

```python
from sklearn.preprocessing import StandardScaler

features = df[["recency", "frequency", "sales"]]

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

scaled_df = pd.DataFrame(scaled_features, columns=features.columns)
scaled_df.head()
```

```python
df.to_csv("../data/processed/processed_data.csv", index=False)
```

### Feature Engineering Insights:
- RFM segmentation helps identify high-value customers
- Scaling improves ML model performance
- Customer segments are now ready for clustering
