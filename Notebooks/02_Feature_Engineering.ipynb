```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score,
    roc_auc_score,
    classification_report,
    confusion_matrix
)

import shap
```

```python
df = pd.read_csv("../data/raw/retail_data.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
required_cols = ["sales", "quantity", "frequency", "recency", "inventory", "churn"]

missing_cols = [col for col in required_cols if col not in df.columns]

if missing_cols:
    raise ValueError(f"Missing columns: {missing_cols}")
```

```python
X = df[["sales", "quantity", "frequency", "recency", "inventory"]]
y = df["churn"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42,
    stratify=y
)
```

```python
model = XGBClassifier(
    n_estimators=200,
    learning_rate=0.05,
    max_depth=4,
    random_state=42,
    eval_metric="logloss"
)

model.fit(X_train, y_train)
```

```python
preds = model.predict(X_test)
probs = model.predict_proba(X_test)[:, 1]

print("Accuracy:", round(accuracy_score(y_test, preds), 4))
print("AUC Score:", round(roc_auc_score(y_test, probs), 4))
print("\nClassification Report:\n", classification_report(y_test, preds))
```

```python
cm = confusion_matrix(y_test, preds)

sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
importance = pd.DataFrame({
    "Feature": X.columns,
    "Importance": model.feature_importances_
}).sort_values("Importance", ascending=False)

importance
```

```python
sns.barplot(data=importance, x="Importance", y="Feature")
plt.title("XGBoost Feature Importance")
plt.show()
```

```python
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X_test)

shap.summary_plot(shap_values, X_test)
```

```python
df["churn_risk"] = model.predict_proba(X)[:, 1]

os.makedirs("../data/processed", exist_ok=True)
df.to_csv("../data/processed/churn_prediction.csv", index=False)

df.sort_values("churn_risk", ascending=False).head(20)
```
