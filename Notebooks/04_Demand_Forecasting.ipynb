```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
# sample time series sales data
dates = pd.date_range(start="2023-01-01", periods=365, freq="D")

import pandas as pd

df = pd.read_csv("../data/raw/retail_data.csv")
df.head()
```

```python
plt.figure(figsize=(10,4))
plt.plot(df["date"], df["sales"])
plt.title("Daily Sales Trend")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()
```

```python
# create moving average forecast
df["forecast"] = df["sales"].rolling(window=7).mean()
df["forecast"] = df["forecast"].fillna(df["sales"].mean())
```

```python
plt.figure(figsize=(10,4))
plt.plot(df["date"], df["sales"], label="Actual Sales")
plt.plot(df["date"], df["forecast"], label="Forecast", color="red")
plt.title("Sales Forecasting")
plt.legend()
plt.show()
```

```python
mae = mean_absolute_error(df["sales"], df["forecast"])
rmse = np.sqrt(mean_squared_error(df["sales"], df["forecast"]))

print("MAE:", mae)
print("RMSE:", rmse)
```

```python
df.to_csv("../data/processed/demand_forecast.csv", index=False)
```
